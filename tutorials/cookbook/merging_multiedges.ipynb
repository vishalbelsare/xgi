```python
import seaborn as sns
import xgi
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
H = xgi.load_xgi_data("diseasome", read=True, path="../../../xgi-data/data/diseasome")
```

```python
H.merge_duplicate_edges(rename="tuple", multiplicity="weight")
```

```python
edge_size = H.edges.size.asnumpy()
multiplicity = H.edges.attrs("weight").asnumpy()
df = pd.DataFrame.from_dict({"edge size": H.edges.size.aslist(), "multiplicity": H.edges.attrs("weight", missing=1).aslist()})
```

```python
g = sns.jointplot(data=df, x="edge size", y="multiplicity", kind="kde")
sns.despine()
# g.ax_joint.set_xscale('log')
# g.ax_joint.set_yscale('log')
```

```python
plt.hist(H.nodes.degree(weight="weight").asnumpy(), bins=10, alpha=0.5, color="red", label="weighted")
plt.hist(H.nodes.degree.asnumpy(), bins=10, alpha=0.5, color="blue", label="unweighted")
plt.legend()
plt.show()
```
